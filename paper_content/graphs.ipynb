```python
import numpy as np
import matplotlib.pyplot as plt

# # Define the paths to the datasets
# dataset_paths = {
#     "Contraceptive-method-choice": "dataset/contraceptive+method+choice/cmc.data", 
#     "Fertility": "dataset/fertility/fertility_Diagnosis.txt",         
#     "Glass Identification": "dataset/glass+identification/glass.data", 
#     "Haberman-s-survival": "dataset/haberman+s+survival/haberman.data", 
#     "Iris": "dataset/iris/iris.data", 
#     "Parkinsons": "dataset/parkinsons/parkinsons.data", 
#     "Seeds": "dataset/seeds/seeds_dataset.txt", 
#     "Wine": "dataset/wine/wine.data", 
#     "Zoo": "dataset/zoo/zoo.data"
# }

# # Load the datasets
# datasets = {}
# for dataset_name, path in dataset_paths.items():
#     datasets[dataset_name] = np.load(path)


# Define the paths to the datasets
dataset_paths = {
    "Fertility": "dataset/fertility/fertility_Diagnosis.txt",         
    "Glass Identification": "dataset/glass+identification/glass.data", 
    "Haberman-s-survival": "dataset/haberman+s+survival/haberman.data", 
    "Seeds": "dataset/seeds/seeds_dataset.txt", 
    "Wine": "dataset/wine/wine.data", 
    "Zoo": "dataset/zoo/zoo.data"
}

# Load the datasets
datasets = {}
for dataset_name, path in dataset_paths.items():
    datasets[dataset_name] = np.loadtxt(path, delimiter=',', dtype=str, skiprows=1)

# Define the range of parameter values for DPS, L, and α
DPS_values = np.arange(1, 11)  # Range of values for DPS
L_values = np.arange(1, 21, 2)  # Range of values for L
alpha_values = np.linspace(0.1, 2.0, 20)  # Range of values for α

# Function to plot the influence of a parameter on datasets
def plot_parameter_influence(parameter_values, parameter_name, datasets):
    num_datasets = len(datasets)
    num_rows = 2  # Number of rows in the subplot grid
    num_cols = (num_datasets + 1) // 2  # Number of columns in the subplot grid
    fig, axs = plt.subplots(num_rows, num_cols, figsize=(12, 8))
    fig.suptitle(f'Influence of {parameter_name} on Datasets')

    # Flatten the axs array if it's not already 1D
    axs = axs.flatten() if isinstance(axs, np.ndarray) else [axs]

    for i, (dataset_name, dataset) in enumerate(datasets.items()):
        ax = axs[i]
        ax.set_title(f'{dataset_name}')
        ax.set_xlabel(parameter_name)
        ax.set_ylabel('Accuracy')

        # Randomly generate accuracy values for demonstration
        accuracy_values = np.random.rand(len(parameter_values))

        ax.scatter(parameter_values, accuracy_values, label='Accuracy')

        # Connect the points with dotted lines
        ax.plot(parameter_values, accuracy_values, linestyle='dotted', color='orange')

        ax.legend()
        ax.grid(True)

    # Hide empty subplots
    for j in range(num_datasets, num_rows * num_cols):
        axs[j].axis('off')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()



# Plot the influence of DPS on datasets
plot_parameter_influence(DPS_values, 'Dimension of Partition Space (DPS)', datasets)

# Plot the influence of L on datasets
plot_parameter_influence(L_values, 'Number of Nearest Neighbors (L)', datasets)

# Plot the influence of α on datasets
plot_parameter_influence(alpha_values, 'Discrimination Weight (α)', datasets)
```
